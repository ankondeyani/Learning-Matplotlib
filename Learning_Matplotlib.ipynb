# **Learning Matplotlib**

Matplotlib a Python Library which can create graphs for us showing the relation between data visually.
This note covers-
*   Creating graphs with different type of plotting like Bar Plot, Pie Chart, Scattered Plot, 3D Scattered Plot
*   Using different type of line style like solid line, dashed line, + sign, dot sign, and many more in a graph
*   Labelling the graph


```python
import matplotlib.pyplot as plt
import numpy as np
```

Importing matplotlib.pyplot as plt is the general convention used by the coders.

```python
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
```

Let's see these datapoints before plotting them.

```python
print(x)
```

```python
print(y)
```

```python
print(z)
```

Now, let's plot them!

```python
plt.plot(x,y)
plt.show()
```

We can also label the graph.

```python
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine values')
plt.title('Sine Wave')
plt.show()
```

## **Line styling**
Let's use different type of colors and line styles inside the graph!

```python
x=np.linspace(-10,10,20)
y=x**2
plt.plot(x,y)
plt.show
```

```python
plt.plot(x,y,'r+')
plt.show
```

```python
plt.plot(x,y,'g.')
plt.show()
```

```python
plt.plot(x,y,'r-')
plt.show()
```

```python
plt.plot(x,y,'g--')
plt.show()
```

# **Various types of Plotting**

## **Bar Graph**

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
lang=['C','C++','Java','Python','JavaScript']
students=[23,17,35,29,12]
ax.bar(lang,students)
plt.xlabel('Languages')
plt.ylabel('No. of Students')
plt.title('Students learning different coding language')
plt.show()
```

##**Pie Chart**

```python
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.pie(students, labels=lang, autopct='%1.1f%%')
plt.title('Students learning different coding language')
plt.show()
```

## **Scatter Plot**

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig3=plt.figure()
ax=fig3.add_axes([0,0,1,1])
ax.scatter(x,y,color='Yellow')
ax.scatter(x,z,color='b')
```

## **3D Scatter Plot**

```python
fig4=plt.figure()
ax=fig4.add_axes([0,0,1,1],projection='3d')
ax.scatter(x,y,z,c=z,cmap='Blues')
```
